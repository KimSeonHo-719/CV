```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

!wget 'https://raw.githubusercontent.com/KimSeonHo-719/CV/main/gyujin.png' -O gyujin.png
!wget 'https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png' -O lena.png

img1_BGR = cv2.imread('/content/gyujin.png')
img2_BGR = cv2.imread('/content/lena.png')

img1=cv2.cvtColor(img1_BGR, cv2.COLOR_BGR2GRAY)
img2=cv2.cvtColor(img2_BGR, cv2.COLOR_BGR2GRAY)

# 두 이미지의 크기가 다를 수 있으므로, 동일한 크기로 리사이즈 (예: img1의 크기에 맞춤)
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# 이미지 확인
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img1,cmap='gray')

plt.subplot(1, 2, 2)
plt.imshow(img2,cmap='gray')

plt.show()
```

```python
# kernel 사이즈
kernel3=3
kernel7=7
kernel15=15

##kernel=3
# 저주파 성분 추출 (cv 라이브러리 내부 구현 함수 - Gaussian Blur)
lp_img1_3=cv2.GaussianBlur(img1,(kernel3,kernel3),0)
lp_img2_3=cv2.GaussianBlur(img2,(kernel3,kernel3),0)

# 고주파 성분 추출
# highpass와 마찬가지로 원본에서 빼면 됨
hp_img1_3=img1-lp_img1_3
hp_img2_3=img2-lp_img2_3

# 이미지1 고주파 성분 + 이미지2 저주파 성분
hybrid1_3=hp_img1_3+lp_img2_3

# 이미지1 저주파 성분 + 이미지2 고주파 성분
hybrid2_3=lp_img1_3+hp_img2_3

plt.imshow(hybrid1_3,cmap='gray')
plt.show()
plt.imshow(hybrid2_3,cmap='gray')
plt.show()
```
